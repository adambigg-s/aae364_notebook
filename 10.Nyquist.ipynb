# Nyquist Plot

```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

$G_c(s) = \dfrac{1}{1 + k G(s)}$

$G_c(s) = \dfrac{1}{1 + k \dfrac{N(s)}{D(s)}}$

$G_c(s) = \dfrac{D(s)}{D(s) + k N(s)}$

zeros of $G_c(s)$ = poles of $G(s)$

```python
s = control.tf([1, 0], [0, 1])
```

```python
G = s/(s+1)
G2 = 10*s/(s+1)
control.bode([G, G2], dB=True);
```

```python
control.rlocus(G);
```

```python
G
```

```python
G
control.nyquist(G, omega=np.logspace(-3, 3));
plt.grid()
```

```python
G2c = G/(1+G)
G2c
```

```python
control.pole(G2c)
```

```python
G = -1/((s+1)*(s+2))
control.bode(G, dB=True);
```

```python
K = 1
control.nyquist(K*G, omega=np.logspace(-2, 2));
```

```python
#Stable, gain margin 20 dB
Gc = K*G/(1+K*G)
```

```python
control.pole(Gc)
```

```python
t, y = control.step_response(Gc, T=np.linspace(0, 1000, 1000));
plt.plot(t, y);
```

```python
G = 1/(s*(s+1))
control.nyquist(G);
Gc = control.minreal(G/(1+G))
control.pole(Gc)
control.stability_margins(G)
```
